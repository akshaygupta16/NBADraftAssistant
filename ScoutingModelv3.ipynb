```python
def get_player_postion(name):
    name.strip()
    full_name=name.split()
    fname=full_name[0].lower()
    lname=full_name[1].lower()
    try:
        sr = 'https://www.sports-reference.com/cbb/players/'+fname+'-'+lname+'-1.html'
        webpage = requests.get(sr)
        docu = lxml.html.fromstring(webpage.content)
        every_p_text = docu.xpath('//p/text()')
        return every_p_text[1].strip()
    except:
        return np.nan
```
